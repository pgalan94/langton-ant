# Langton's Ant implementation with OOP

## Class Definition


```python
SQUARE_SIZE = 5
BOARD_SIZE = 199
ANT_SIZE = 2
```

```python
class Square:
    def __init__(self, size, px, py):
        self.size = size
        self.color = 'white'
        self.px = px
        self.py = py

    def toggle_color(self):
        if self.color == 'white':
            self.color = 'black'
            return
        self.color = 'white'

    def __mul__(self, other):
        if isinstance(other, int):
            return [self for _ in range(other)]
        else:
            raise NotImplemented

    def rect(self):
        return '<rect x="{px}" y="{py}" width="{x}" height="{x}" stroke="black" stroke-width="1" fill="{c}"/>'.format(px=self.px, py=self.py, x=self.size, c=self.color)
```

```python
class Board:
    def __init__(self, size):
        self.size = size
        self.grid = {}
        for i in range(size):
            self.grid[i] = {}
            for j in range(size):
                self.grid[i][j] = Square(SQUARE_SIZE, SQUARE_SIZE*j, SQUARE_SIZE*i)

    def svg(self, ant):
        return """
        <svg width='{x}' height='{x}' style='background: #f7f7f9'>
        """.format(x=self.size * SQUARE_SIZE) + """
        {}
        </svg>""".format("\n".join("".join(v.rect() for v in self.grid[row].values()) for row in self.grid.keys()) + ant.svg())

    def draw(self):
        pass
```

```python
class Ant:
    def __init__(self, size, speed, x, y):
        self.size = size
        self.speed = speed
        self.x = x
        self.y = y
        self.direction = 1

    def svg(self):
        return '<circle cx="{px}" cy="{py}" r="{s}" stroke="black" stroke-width="1" fill="red" />'.format(px=self.x*SQUARE_SIZE - SQUARE_SIZE/2, py=self.y*SQUARE_SIZE - SQUARE_SIZE/2, s=self.size)


    def langton_direction(self, board):
        if board.grid[self.y - 1][self.x - 1].color == 'white':
            self.direction += 1
            if self.direction > 4:
                self.direction = 1
            return
        self.direction -= 1
        if self.direction < 1:
            self.direction = 4

    def paint_square(self, board, x, y):
        board.grid[self.y - 1][self.x - 1].toggle_color()

    def move(self, board):
        self.langton_direction(board)
        if self.direction == 1:
            self.x += 1
        if self.direction == 2:
            self.y += 1
        if self.direction == 3:
            self.x -= 1
        if self.direction == 4:
            self.y -= 1
        if self.x < 1:
            self.x = BOARD_SIZE
        if self.y < 1:
            self.y = BOARD_SIZE
        if self.x > BOARD_SIZE:
            self.x = 1
        if self.y > BOARD_SIZE:
            self.y = 1
```

```python
board = Board(BOARD_SIZE)
ant = Ant(ANT_SIZE, 1, BOARD_SIZE//2, BOARD_SIZE//2)
```

```python
from IPython.display import clear_output, display, HTML

for i in range(100000):
    clear_output(wait=True)
    last_square_x = ant.x
    last_square_y = ant.y
    ant.move(board)
    
    print(i)
    ant.paint_square(board, last_square_x, last_square_y)
```
